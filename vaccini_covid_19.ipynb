importa le librerie necessarie per il funzionamento del codice

```python
import pandas as pd
import plotly.graph_objects as go
```

estrai le informazioni sui vaccini con pandas

```python
nome_file="somministrazioni-vaccini-latest.csv"

df = pd.read_csv(nome_file)

data_somministrazione = df.data_somministrazione.to_list()
fornitore = df.fornitore.to_list()
area = df.area.to_list()
fascia_anagrafica = df.fascia_anagrafica.to_list()
sesso_maschile = df.sesso_maschile.to_list()
sesso_femminile = df.sesso_femminile.to_list()
categoria_operatori_sanitari_sociosanitari = df.categoria_operatori_sanitari_sociosanitari.to_list()
categoria_personale_non_sanitario = df.categoria_personale_non_sanitario.to_list()
categoria_ospiti_rsa = df.categoria_ospiti_rsa.to_list()
categoria_over80 = df.categoria_over80.to_list()
categoria_personale_scolastico = df.categoria_personale_scolastico.to_list()
categoria_altro = df.categoria_altro.to_list()
prima_dose = df.prima_dose.to_list()
seconda_dose = df.seconda_dose.to_list()
codice_NUTS1 = df.codice_NUTS1.to_list()
codice_NUTS2 = df.codice_NUTS2.to_list()
codice_regione_ISTAT = df.codice_regione_ISTAT.to_list()
nome_area = df.nome_area.to_list()
```

informazioni sulla somministrazione

```python
totale_prima_dose = (sum(prima_dose))
totale_seconda_dose = (sum(seconda_dose))

print("\ntotale prima dose",totale_prima_dose/1000000,"milioni")
print("totale seconda dose (vaccinati)",totale_seconda_dose/1000000,"milioni")
print("totale dosi somministrate (prima e seconda)",(totale_prima_dose+totale_seconda_dose)/1000000,"milioni")

print ("\npercentuale prima dose su 60 milioni",totale_prima_dose*100/60000000)
print ("percentuale seconda dose su 60 milioni",totale_seconda_dose*100/60000000)
```

grafico somministrazione prima dose

```python
fig = go.Figure(data=[
    go.Bar(name='data_somministrazione', x=data_somministrazione, y=prima_dose)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```

grafico somministrazione seconda dose

```python
fig = go.Figure(data=[
    go.Bar(name='data_somministrazione', x=data_somministrazione, y=seconda_dose)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
